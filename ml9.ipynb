```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
path ="/content/drive/MyDrive/ML_Dataset/Mall_Customers.csv"
df=pd.read_csv(path)
df.head(5)
```

```python
X=df.iloc[:,[3,4]].values
print(X)
```

```python
#feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#optimal clusters
inertia_values = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
  kmeans.fit(X_scaled)
  inertia_values.append(kmeans.inertia_)
print(inertia_values)
plt.figure()
plt.plot(range(1,11), inertia_values, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
#elbow method tells k konsa choose krna h
```

```python
from numpy.random.mtrand import random
kmeans = KMeans(n_clusters=5,random_state=42)
clusters = kmeans.fit_predict(X_scaled)

df['Cluster'] = clusters
print(df.head())
```

```python
print("Inertia",kmeans.inertia_)
sil_score = silhouette_score(X_scaled,clusters)
print("Silhouette Score",sil_score)
```

```python
plt.figure()
sns.scatterplot(x=df['Annual Income (k$)'],y='Spending Score (1-100)',hue='Cluster',data=df,palette='Set1')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Clusters')

#centroids

centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(
    centroids[:,0],
    centroids[:,1],
    s=200,
    marker='X',
    color='black',
    label='Centroids'
)

plt.title('Customer Segmetation using K-Means')
plt.legend()
plt.show()
```
